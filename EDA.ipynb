```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
import nibabel as nib
import pandas as pd
```

```python
##Load the .nii files in the folder
files = glob.glob('./data/**/*.nii',recursive=True)
```

```python
#put all the image files in a dictionary
imageFiles = {file.split('/')[2]:[] for file in files}
for file in files:
    imageFiles[file.split('/')[2]].append(file)
```

```python
patientDF=pd.DataFrame(columns=['ID:No','DWI Max','DWI Min','Flair Max','Flair Min','T1 Max','T1 Min','T2 Max','T2 Min','X Shape','Y Shape','Z Shape'])
```

```python
patients=sorted([int(x) for x in list(imageFiles.keys())])
```

```python
for idno in patients:
    dataframe = []
    dataframe.append(int(idno))
    for file in sorted(imageFiles[str(idno)])[:-1]:
        img = nib.load(file).get_fdata()
        dataframe.append(img.max())
        dataframe.append(img.min())
    dataframe.append(img.shape[0])
    dataframe.append(img.shape[1])
    dataframe.append(img.shape[2])
    patientDF.loc[len(patientDF)]=dataframe
        
```

```python
def getDetails(imageFiles):
    patientDF=pd.DataFrame(columns=['patient','Grey Matter Volume', 'Stroke Volume', 'percentage affected'])
    patients=sorted([int(x) for x in list(imageFiles.keys())])
    for patient in patients:
        
        scanPath=sorted(imageFiles[str(patient)])[-2]
        scan=nib.load(scanPath).get_fdata()
        
        # normalise 
        scan-=scan.min()
        scan/=scan.max()
        
        # count all non zeros values get brain values
        greyMatterVolume=np.count_nonzero(scan)
        annotPath=sorted(imageFiles[str(patient)])[-1]
        
        annot=nib.load(annotPath).get_fdata()
        
        # count all non zeros values get storke values
        strokeVolume=np.count_nonzero(annot)

        entry=[patient,greyMatterVolume,strokeVolume,(strokeVolume/greyMatterVolume)*100]
        
        patientDF.loc[len(patientDF)]=entry
    
    return patientDF     
    
```

```python
StrokeDF=getDetails(imageFiles)
```

```python
def visualizeStroke(view,images=5):
    path='./dataPreprocessed/'+view
    
    views=['sagital','frontal','coronal']
    files=glob.glob(path+'/*_*')
    files+=[path+'/annot']
    # load the modalities for the view
    listScans=[np.load(file,allow_pickle=True) for file in files]
    
    ## get random locations to display
    randomLoc=np.random.randint(listScans[0].shape[views.index(view)], size=images)
    
    for loc in randomLoc:
        fig=plt.figure(figsize=(20,5))
        seg=listScans[-1]
        for i, img in enumerate(listScans[:-1]):
            plt.subplot(1,4,i+1)
            plt.title(files[i].split('/')[-1])
            if view=='coronal':
                # plot the image
                plt.imshow(np.rot90(img[:,:,loc]),cmap='gray')
                # overlay the segmentation mask
                plt.imshow(np.rot90(seg[:,:,loc]),alpha=0.4,cmap='Blues')
            elif view=='sagital':
                # plot the image
                plt.imshow(np.rot90(img[loc,:,:]),cmap='gray')
                # overlay the segmentation mask
                plt.imshow(np.rot90(seg[loc,:,:]),alpha=0.4,cmap='Blues')
            elif view=='frontal':
                # plot the image
                plt.imshow(np.rot90(img[:,loc,:]),cmap='gray')
                # overlay the segmentation mask
                plt.imshow(np.rot90(seg[:,loc,:]),alpha=0.4,cmap='Blues')
                

            plt.axis('off')
        plt.show()
```

```python
visualizeStroke('coronal')
```
